```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.array([0, 0.25, 0.5, 0.75, 1])
prior = np.array([1, 1, 1, 1, 1])/5

samples = np.array([1, 0, 1, 1, 1, 0, 1, 0, 1])

fig, axs = plt.subplots(3, 3, figsize=(13, 10))
l = 0
m = 0

for i in samples:
    prob_data = i * p + (1 - i) * (1 - p)
    posterior = prior * prob_data
    posterior_n = posterior/np.sum(posterior)
    print(posterior_n)
    axs[l, m].plot(p, prior, ls = '--')
    axs[l, m].plot(p, posterior_n)
    axs[l, m].grid(color = 'grey', linestyle = '--', linewidth = 0.5)
    prior = posterior_n
    m += 1
    if m % 3 == 0:
        m = 0
        l += 1

fig.tight_layout(pad=1.5)
plt.show()
```
